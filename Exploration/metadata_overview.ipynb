```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Read document metadata table and summarise
df_meta = pd.read_csv('metadata_sample.csv')
display(df_meta.info())
display(df_meta.head())
df_meta.tail()
```

```python
# Generate frequency graphs (documents as unit of analysis)
title_dict = {'ecco_pages':'Number of Page', 
              'language_primary': 'Primary Language',
              'gatherings':'Gathering Type', 
              'publication_year': 'Publication Year', 
              'publication_decade': 'Publication Decade', 
              'document_type': 'Document Type',
              'work_id': 'Name of Work (Abbrev.)',
              'publication_place': 'Publication Place', 
              'publication_country': 'Publication Country',
              'authors': 'Author',
              'sub_category': 'Category'}

for item in ['ecco_pages','language_primary', 'gatherings', 'publication_year', 'publication_decade', 'document_type']:
    print("---------------------------")
    freq_dict = dict(df_meta[item].value_counts())
    plt.bar(freq_dict.keys(), freq_dict.values())
    plt.title(f'Documents (Books) per {title_dict[item]}')
    plt.xlabel(item)
    plt.ylabel('frequency')
    #plt.savefig(f"{item}.png" , bbox_inches='tight')
    plt.show()

for item in ['work_id', 'publication_place', 'publication_country', 'authors']:
    print("---------------------------")
    freq_dict = dict(df_meta[item].value_counts()[:10])
    plt.title(f'Documents (Books) per {title_dict[item]}')
    plt.bar(freq_dict.keys(), freq_dict.values())
    plt.xlabel(item)
    plt.xticks(rotation = 90)
    plt.ylabel('frequency')
    #plt.savefig(f"{item}.png", bbox_inches='tight')
    plt.show()

#'sub_category'
print("---------------------------")
plt.title('Documents (Books) per Category')
freq_dict = dict(df_meta['sub_category'].value_counts()[:10])
plt.bar(freq_dict.keys(), freq_dict.values())
plt.xlabel(title_dict['sub_category'])
plt.xticks(rotation = 90)
plt.ylabel('frequency')
#plt.savefig(f"sub_category.png", bbox_inches='tight')
plt.show()
```

```python
# read the illustration table and summarise
df_illu = pd.read_csv('ecco_sci_urls.csv')
display(df_illu.info())
display(df_illu.head())
df_illu.tail()
```

```python
# Map the illustration with document metadata 
df = df_illu.merge(df_meta, on='ecco_id')

# Generate frequency graphs (illustrations as unit of analysis)
for item in ['ecco_pages','language_primary', 'gatherings', 'publication_year', 'publication_decade', 'document_type']:
    print("---------------------------")
    freq_dict = dict(df[item].value_counts())
    plt.bar(freq_dict.keys(), freq_dict.values(), color='crimson')
    plt.title(f'Illustrations per {title_dict[item]}')
    plt.xlabel(item)
    plt.ylabel('frequency')
    #plt.savefig(f"{item}.png", bbox_inches='tight')
    plt.show()

for item in ['work_id', 'publication_place', 'publication_country', 'authors']:
    print("---------------------------")
    freq_dict = dict(df[item].value_counts()[:10])
    plt.title(f'Illustrations per {title_dict[item]}')
    plt.bar(freq_dict.keys(), freq_dict.values(), color='crimson')
    plt.xlabel(item)
    plt.xticks(rotation = 90)
    plt.ylabel('frequency')
    #plt.savefig(f"{item}.png", bbox_inches='tight')
    plt.show()

#'sub_category'
print("---------------------------")
plt.title('Illustrations per Category')
freq_dict = dict(df['sub_category'].value_counts()[:10])
plt.bar(freq_dict.keys(), freq_dict.values(), color='crimson')
plt.xlabel('sub_category')
plt.xticks(rotation = 90)
plt.ylabel('frequency')
#plt.savefig("sub_category.png", bbox_inches='tight')
plt.show()
```

```python
# Proportion of illustrations (on total number of pages)

n_img = df.groupby(by='publication_year').size().reset_index(name='img_count')
n_p = df_meta[['ecco_id','publication_year', 'ecco_pages']].drop_duplicates().groupby(by='publication_year')[['ecco_pages']].sum().reset_index()

plt.figure(figsize=(10,6), tight_layout=True)
plt.bar(n_img.publication_year, n_img.img_count/n_p.ecco_pages, alpha=1)
plt.xlabel("Year")
plt.ylabel('Proportion')
plt.title('Illustrations per page by Publication Year')
plt.savefig('illustration_per_page.png')
```

```python
# Looking at 10 most important document categories (with most illustrations)
df['sub_category'] = df['sub_category'].str.replace('scientificimprovement','')
cats=df.groupby("sub_category").size().sort_values(ascending=False).head(10)
display(cats)

subdf=df[df["sub_category"].isin(cats.index)]
print('\nSummary of the data frame with only 10 most important categories')
subdf.info()
```

```python
# How the number of documents in each top-10 category changed over decades?

sns.set_theme(style="whitegrid",palette="PuOr_r")

dec=subdf.groupby(["sub_category","publication_decade"])[["estc_id","ecco_id"]].nunique().reset_index()

display(dec.head())

# Plot
fig, ax= plt.subplots(1,1,figsize=(8,6))

ax=sns.scatterplot(data=dec[dec["publication_decade"]<1800], x ="publication_decade", y="ecco_id", hue="sub_category")

ax=sns.lineplot(data=dec[dec["publication_decade"]<1800], x ="publication_decade", y="ecco_id", hue="sub_category")

#plt.margins(0,0)

plt.xlabel("Decade")
plt.ylabel("Number of unique ESTC IDs")

h,l = ax.get_legend_handles_labels()
plt.legend(h[0:10],l[0:10],loc='upper left',#, bbox_to_anchor=(1.18, 
                             #                  0.5),
          ncol=1, fancybox=False, shadow=False)

#ax.xaxis.set_tick_params(rotation=0)
fig.tight_layout()
# Turns off grid on the left Axis.
ax.xaxis.grid([])
#fig.savefig("dec_category.png")
```

```python
# How the proportion of illustration in each top-10 category changed over decades?

p_dec = subdf.drop_duplicates('ecco_id').groupby(["sub_category","publication_decade"])['ecco_pages'].sum().reset_index()
ill_dec= subdf.groupby(["sub_category","publication_decade"]).size().reset_index(name="img_count")
ill_dec = ill_dec.merge(p_dec, on=["sub_category","publication_decade"])
display(ill_dec.head())
ill_dec['illu_prop'] = ill_dec.apply(lambda x: x['img_count']/x['ecco_pages'], axis = 1)


fig, ax= plt.subplots(1,1,figsize=(8,6))

ax=sns.scatterplot(data=ill_dec[ill_dec["publication_decade"]<1800], x ="publication_decade", y="illu_prop", hue="sub_category")

ax=sns.lineplot(data=ill_dec[ill_dec["publication_decade"]<1800], x ="publication_decade", y="illu_prop", hue="sub_category")

#plt.margins(0,0)

plt.xlabel("Decade")
plt.ylabel("Proportion of Pictures")

h,l = ax.get_legend_handles_labels()
plt.legend(h[0:10],l[0:10],loc='upper left', #bbox_to_anchor=(1.18, 0.5),
          ncol=1, fancybox=False, shadow=False)

#ax.xaxis.set_tick_params(rotation=0)
fig.tight_layout()
# Turns off grid on the left Axis.
ax.xaxis.grid([])
#plt.savefig('proportion_of_pictures.png')
```
