```python
low_lst = dfL.Image.tolist()
high_lst = dfH.Image.tolist()
cheap_lst = c_df.values.tolist()
exp_lst = e_df.values.tolist()
```

```python
cnt = 0
CresultsL = {}
CresultsH = {}

for s in cheap_lst:
    id = "{:010d}".format(s[0])
    if id not in CresultsL.keys() and id not in CresultsH.keys():
        CresultsL[id] = []
        CresultsH[id] = []
    for x in low_lst:
        if x.startswith(id) and id in CresultsL.keys():
            CresultsL[id].append(x)
    for x in high_lst:
        if x.startswith(id) and id in CresultsH.keys():
            CresultsH[id].append(x)
```

```python
CresultsL
```

```python
cnt = 0
EresultsL = {}
EresultsH = {}

for s in exp_lst:
    id = "{:010d}".format(s[0])
    if id not in EresultsL.keys() and id not in EresultsH.keys():
        EresultsL[id] = []
        EresultsH[id] = []
    for x in low_lst:
        if x.startswith(id) and id in EresultsL.keys():
            EresultsL[id].append(x)
    for x in high_lst:
        if x.startswith(id) and id in EresultsH.keys():
            EresultsH[id].append(x)
```

```python
Cresults = {}
for t1,t2 in zip(CresultsH.keys(), CresultsL.keys()):
    
    if len(CresultsH[t1]) > len(CresultsL[t2]):
        Cresults[t1] = "High"
    else:
        Cresults[t1] = "Low"

Eresults = {}
for t1,t2 in zip(EresultsH.keys(), EresultsL.keys()):
    
    if len(EresultsH[t1]) > len(EresultsL[t2]):
        Eresults[t1] = "High"
    else:
        Eresults[t1] = "Low"
```

```python
CyearsL = {}
CyearsH = {}

for book in Cresults.keys():
    try:
        if Cresults[book] == "Low":
            
            if int(c_df[c_df.ecco_id2 == book].publication_year) not in CyearsL.keys():
                CyearsL[int(c_df[c_df.ecco_id2 == book].publication_year)] = [book]
            else:
                CyearsL[int(c_df[c_df.ecco_id2 == book].publication_year)].append(book)
        else:
            if int(c_df[c_df.ecco_id2 == book].publication_year) not in CyearsH.keys():
                CyearsH[int(c_df[c_df.ecco_id2 == book].publication_year)] = [book]
            else:
                CyearsH[int(c_df[c_df.ecco_id2 == book].publication_year)].append(book)
    except:
        continue
```

```python
EyearsL = {}
EyearsH = {}

for book in Eresults.keys():
    try:
        if Eresults[book] == "Low":
            
            if int(e_df[e_df.ecco_id2 == book].publication_year) not in EyearsL.keys():
                EyearsL[int(e_df[e_df.ecco_id2 == book].publication_year)] = [book]
            else:
                EyearsL[int(e_df[e_df.ecco_id2 == book].publication_year)].append(book)
        else:
            if int(e_df[e_df.ecco_id2 == book].publication_year) not in EyearsH.keys():
                EyearsH[int(e_df[e_df.ecco_id2 == book].publication_year)] = [book]
            else:
                EyearsH[int(e_df[e_df.ecco_id2 == book].publication_year)].append(book)
    except:
        continue
```

```python
def getBooksfromYear(year, results):
    if year not in results.keys():
        return []
    else:
        return results[year]
```

```python
def drawPics(EvolumeL, EvolumeH, CvolumeL,CvolumeH):
    figure, axis = plt.subplots(1, 2,figsize=(16,6), sharex=True, sharey=True)
    #figure.suptitle("Quality of illustrations in 18th century")

    axis[0].set_title('#Prestigious books (2fo,4to)')
    axis[0].grid(True)

    xL = [p[0] for p in EvolumeL]
    yL = [p[1] for p in EvolumeL]
    xH = [p[0] for p in EvolumeH]
    yH = [p[1] for p in EvolumeH]
    axis[0].scatter(xH,yH, c="b", label="High quality")
    axis[0].scatter(xL,yL, c="r", label="Low quality")
    zH = np.polyfit(xH, yH, 1)
    zL = np.polyfit(xL, yL, 1)
    pL = np.poly1d(zL)
    pH = np.poly1d(zH) 

    axis[0].plot(xL, pL(xL), "r")
    axis[0].plot(xH, pH(xH), "b")
    axis[0].legend()

    axis[1].set_title('#Popular books (8vo,12mo)')
    axis[1].grid(True)

    xL = [p[0] for p in CvolumeL]
    yL = [p[1] for p in CvolumeL]
    xH = [p[0] for p in CvolumeH]
    yH = [p[1] for p in CvolumeH]
    axis[1].scatter(xH,yH, c="b", label="High quality")
    axis[1].scatter(xL,yL, c="r", label="Low quality")
    zH = np.polyfit(xH, yH, 1)
    zL = np.polyfit(xL, yL, 1)
    pL = np.poly1d(zL)
    pH = np.poly1d(zH) 

    axis[1].plot(xL, pL(xL), "r")
    axis[1].plot(xH, pH(xH), "b")
#    axis[1].legend()



    figure.savefig("C:/Users/arive/DHH23/quality.png")
```
